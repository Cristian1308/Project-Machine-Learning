Historiagramas o detección de intervalos

Se importan las librerías necesarias para análisis de datos y visualización.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Se carga el dataset desde Google Drive.

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Exploratorio/card_transdata.csv")
```

Se separan las transacciones en dos grupos:

```python
df_fraude = df[df["fraud"] == 1]
df_legitimo = df[df["fraud"] == 0]

df_fraude_sample = df_fraude.sample(n=10000, random_state=1) if len(df_fraude) > 10000 else df_fraude
df_legitimo_sample = df_legitimo.sample(n=10000, random_state=1) if len(df_legitimo) > 10000 else df_legitimo
```

Se define una lista de variables numéricas relevantes para visualizar su distribución.

```python
variables_relevantes = [
    "distance_from_home",
    "distance_from_last_transaction",
    "ratio_to_median_purchase_price"
]
```

Se configura el estilo de los gráficos y se crea un espacio con 3 subgráficos en una sola fila.

Se genera un histograma para cada variable, comparando la distribución entre transacciones legítimas (azul) y fraudulentas (rojo).

Se ajusta el diseño para evitar superposición y se muestra el gráfico final.

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(ncols=3, figsize=(20, 5))

for i, variable in enumerate(variables_relevantes):
    sns.histplot(df_legitimo[variable], label="Legítima", kde=False, color="blue", ax=axes[i], stat="density")
    sns.histplot(df_fraude[variable], label="Fraude", kde=False, color="red", ax=axes[i], stat="density")
    axes[i].set_title(f"Distribución de '{variable}'")
    axes[i].legend(loc="upper right")

plt.tight_layout()
```
